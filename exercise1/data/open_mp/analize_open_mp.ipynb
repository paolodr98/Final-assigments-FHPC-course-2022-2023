## Notebook for plots ex 1 - OpenMP scalability

```python
# Import the libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import glob
import os
```

## Import the data 

```python
# -----------------STATIC----------------------------------------
path_THIN_static_1 = 'static/THIN_static_1.csv'
path_THIN_static_2 = 'static/THIN_static_2.csv'
path_THIN_static_3 = 'static/THIN_static_3.csv'
path_THIN_static_4 = 'static/THIN_static_4.csv'


THIN_static_1 = pd.read_csv(path_THIN_static_1)
THIN_static_2 = pd.read_csv(path_THIN_static_2)
THIN_static_3 = pd.read_csv(path_THIN_static_3)
THIN_static_4 = pd.read_csv(path_THIN_static_4)

# -----------------ORDERED----------------------------------------
path_THIN_ordered_1 = 'ordered/THIN_ordered_1.csv'
path_THIN_ordered_2 = 'ordered/THIN_ordered_2.csv'

THIN_ordered_1 = pd.read_csv(path_THIN_ordered_1)
THIN_ordered_2 = pd.read_csv(path_THIN_ordered_2)


```

## Manipulate the data

```python
# extract the first row for the x axis

n_threads = THIN_static_1['threads']
n_threads = sorted(n_threads.unique())
```

```python
# MPI 1

THIN_static_1_means = THIN_static_1.groupby(['ksize', 'threads']).mean().to_numpy()
THIN_static_1_std = THIN_static_1.groupby(['ksize', 'threads']).std().to_numpy()

# MPI 2

THIN_static_2_means = THIN_static_2.groupby(['ksize', 'threads']).mean().to_numpy()
THIN_static_2_std = THIN_static_2.groupby(['ksize', 'threads']).std().to_numpy()

# MPI 3

THIN_static_3_means = THIN_static_3.groupby(['ksize', 'threads']).mean().to_numpy()
THIN_static_3_std = THIN_static_3.groupby(['ksize', 'threads']).std().to_numpy()

# MPI 4

THIN_static_4_means = THIN_static_4.groupby(['ksize', 'threads']).mean().to_numpy()
THIN_static_4_std = THIN_static_4.groupby(['ksize', 'threads']).std().to_numpy()


#--------------ORDERED---------------


# MPI 1

THIN_ord_1_means = THIN_ordered_1.groupby(['ksize', 'threads']).mean().to_numpy()
THIN_ord_1_std = THIN_ordered_1.groupby(['ksize', 'threads']).std().to_numpy()

# MPI 2

THIN_ord_2_means = THIN_ordered_2.groupby(['ksize', 'threads']).mean().to_numpy()
THIN_ord_2_std = THIN_ordered_2.groupby(['ksize', 'threads']).std().to_numpy()




```

```python
# MPI 1

THIN_static_1_means_ksize_100 = THIN_static_1_means[0:7].reshape(7)
THIN_static_1_means_ksize_500 = THIN_static_1_means[7:14].reshape(7)
THIN_static_1_means_ksize_1000 = THIN_static_1_means[14:21].reshape(7)
THIN_static_1_means_ksize_5000 = THIN_static_1_means[21:28].reshape(7)
THIN_static_1_means_ksize_10000 = THIN_static_1_means[28:35].reshape(7)

THIN_static_1_std_ksize_100 = THIN_static_1_std[0:7].reshape(7)
THIN_static_1_std_ksize_500 = THIN_static_1_std[7:14].reshape(7)
THIN_static_1_std_ksize_1000 = THIN_static_1_std[14:21].reshape(7)
THIN_static_1_std_ksize_5000 = THIN_static_1_std[21:28].reshape(7)
THIN_static_1_std_ksize_10000 = THIN_static_1_std[28:35].reshape(7)

# MPI 2

THIN_static_2_means_ksize_100 = THIN_static_2_means[0:7].reshape(7)
THIN_static_2_means_ksize_500 = THIN_static_2_means[7:14].reshape(7)
THIN_static_2_means_ksize_1000 = THIN_static_2_means[14:21].reshape(7)
THIN_static_2_means_ksize_5000 = THIN_static_2_means[21:28].reshape(7)
THIN_static_2_means_ksize_10000 = THIN_static_2_means[28:35].reshape(7)

THIN_static_2_std_ksize_100 = THIN_static_2_std[0:7].reshape(7)
THIN_static_2_std_ksize_500 = THIN_static_2_std[7:14].reshape(7)
THIN_static_2_std_ksize_1000 = THIN_static_2_std[14:21].reshape(7)
THIN_static_2_std_ksize_5000 = THIN_static_2_std[21:28].reshape(7)
THIN_static_2_std_ksize_10000 = THIN_static_2_std[28:35].reshape(7)

# MPI 3

THIN_static_3_means_ksize_100 = THIN_static_3_means[0:7].reshape(7)
THIN_static_3_means_ksize_500 = THIN_static_3_means[7:14].reshape(7)
THIN_static_3_means_ksize_1000 = THIN_static_3_means[14:21].reshape(7)
THIN_static_3_means_ksize_5000 = THIN_static_3_means[21:28].reshape(7)
THIN_static_3_means_ksize_10000 = THIN_static_3_means[28:35].reshape(7)

THIN_static_3_std_ksize_100 = THIN_static_3_std[0:7].reshape(7)
THIN_static_3_std_ksize_500 = THIN_static_3_std[7:14].reshape(7)
THIN_static_3_std_ksize_1000 = THIN_static_3_std[14:21].reshape(7)
THIN_static_3_std_ksize_5000 = THIN_static_3_std[21:28].reshape(7)
THIN_static_3_std_ksize_10000 = THIN_static_3_std[28:35].reshape(7)


# MPI 4

THIN_static_4_means_ksize_100 = THIN_static_4_means[0:7].reshape(7)
THIN_static_4_means_ksize_500 = THIN_static_4_means[7:14].reshape(7)
THIN_static_4_means_ksize_1000 = THIN_static_4_means[14:21].reshape(7)
THIN_static_4_means_ksize_5000 = THIN_static_4_means[21:28].reshape(7)
THIN_static_4_means_ksize_10000 = THIN_static_4_means[28:35].reshape(7)

THIN_static_4_std_ksize_100 = THIN_static_4_std[0:7].reshape(7)
THIN_static_4_std_ksize_500 = THIN_static_4_std[7:14].reshape(7)
THIN_static_4_std_ksize_1000 = THIN_static_4_std[14:21].reshape(7)
THIN_static_4_std_ksize_5000 = THIN_static_4_std[21:28].reshape(7)
THIN_static_4_std_ksize_10000 = THIN_static_4_std[28:35].reshape(7)


#--------------ORDERED---------------


# MPI 1

THIN_ord_1_means_ksize_100 = THIN_ord_1_means[0:7].reshape(7)
THIN_ord_1_means_ksize_500 = THIN_ord_1_means[7:14].reshape(7)
THIN_ord_1_means_ksize_1000 = THIN_ord_1_means[14:21].reshape(7)
THIN_ord_1_means_ksize_5000 = THIN_ord_1_means[21:28].reshape(7)
THIN_ord_1_means_ksize_10000 = THIN_ord_1_means[28:35].reshape(7)

THIN_ord_1_std_ksize_100 = THIN_ord_1_std[0:7].reshape(7)
THIN_ord_1_std_ksize_500 = THIN_ord_1_std[7:14].reshape(7)
THIN_ord_1_std_ksize_1000 = THIN_ord_1_std[14:21].reshape(7)
THIN_ord_1_std_ksize_5000 = THIN_ord_1_std[21:28].reshape(7)
THIN_ord_1_std_ksize_10000 = THIN_ord_1_std[28:35].reshape(7)

# MPI 2

THIN_ord_2_means_ksize_100 = THIN_ord_2_means[0:7].reshape(7)
THIN_ord_2_means_ksize_500 = THIN_ord_2_means[7:14].reshape(7)
THIN_ord_2_means_ksize_1000 = THIN_ord_2_means[14:21].reshape(7)

THIN_ord_2_std_ksize_100 = THIN_ord_2_std[0:7].reshape(7)
THIN_ord_2_std_ksize_500 = THIN_ord_2_std[7:14].reshape(7)
THIN_ord_2_std_ksize_1000 = THIN_ord_2_std[14:21].reshape(7)


```
